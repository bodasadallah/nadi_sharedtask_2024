```python
dialects = ['MSA', 'Cairo', 'Riyadh', 'Aleppo', 'Amman', 'Beirut', 'Damascus', 'Doha', 'Jeddah', 'Jerusalem', 'Muscat', 'Salt']

dias = {}
for dialect in dialects:
    dia_df = pd.read_csv (f"data/MADAR_Corpus/MADAR.corpus.{dialect}.tsv", sep = '\t')
    print (dia_df)
    dias[dialect] = dia_df
    

```

```python
dias['MSA'].head()
```

```python
big_df = ""
for dia in dialects:
    if dia != 'MSA':
        dias[dia] = dias[dia].merge(dias['MSA'], how='left', suffixes=("",f"_MSA"), on = 'sentID.BTEC')
```

```python
dias['Aleppo']
```

```python
dialects = ['MSA', 'Cairo', 'Riyadh', 'Aleppo', 'Amman', 'Beirut', 'Damascus', 'Doha', 'Jeddah', 'Jerusalem', 'Muscat', 'Salt']
original = 'Egyptian, Emariti, Jordanian, and Palestinian'

mapper = {

    'Egyptian': ['Cairo'],
    'Emariti': ['Doha','Riyadh','Jeddah','Muscat'],
    'Jordanian': ['Amman','Salt'],
    'Palestinian': ['Jerusalem','Aleppo','Damascus','Beirut'],
}


for dia in dialects:
    if not dia == 'MSA':
        dias[dia] = dias[dia][['sent', 'sent_MSA']]

        for key in mapper.keys():
            if dia in mapper[key]:
                dias[dia]['dialect'] = key
```

```python
for dia in dialects:
    if not dia == 'MSA':
        dias[dia] = dias[dia].rename(columns = {'sent':'source', 'sent_MSA':'target' })


test_data = test_data.rename(columns = {'source_da':'source', 'target_msa'   :'target', 'dialect_id':'dialect'})
dev_data = dev_data.rename(columns = {'source_da':'source', 'target_msa'   :'target', 'dialect_id':'dialect'})
```

```python
test_data.columns, dev_data.columns, dias['Cairo'].columns
```

```python
gpt_data_egypt = pd.read_csv('data/MSA_and_Egyptian.csv')
gpt_data_lebnese = pd.read_csv('data/MSA_and_Lebanese.csv')
gpt_data_saudi = pd.read_csv('data/MSA_and_Saudi.csv')
gpt_data_syrian = pd.read_csv('data/MSA_and_Syrian.csv')
```

```python
gpt_data_lebnese.head()
```

```python
gpt_data_egypt = gpt_data_egypt.rename(columns = {'MSA':'source', 'Egyptian':'target', 'lang':'dialect'})
gpt_data_lebnese = gpt_data_lebnese.rename(columns = {'MSA':'source', 'Lebanese':'target', 'lang':'dialect'})
gpt_data_saudi = gpt_data_saudi.rename(columns = {'MSA':'source', 'Saudi':'target', 'lang':'dialect'})
gpt_data_syrian = gpt_data_syrian.rename(columns = {'MSA':'source', 'Syrian':'target', 'lang':'dialect'})

```

```python
gpt_data_lebnese['dialect']  = 'Palestinian'
gpt_data_syrian['dialect']  = 'Palestinian'
gpt_data_saudi['dialect']  = 'Emariti'
gpt_data_egypt['dialect']  = 'Egyptian'
```

```python
train_df = pd.concat([gpt_data_egypt, gpt_data_lebnese, gpt_data_saudi, gpt_data_syrian] + list(dias.values()) )
```

```python
train_df = train_df[['source', 'target', 'dialect']]
```

```python
train_df.reset_index(drop=True, inplace=True)
```

```python
train_df
```

```python
train_df = train_df.dropna()
```

```python
train_df
```

```python
dev_data[dev_data['dialect'].str.contains('Egyptain')] = 'Egyptian'
```

```python
from datasets import DatasetDict

train_dataset = Dataset.from_pandas(train_df)
dev_dataset = Dataset.from_pandas(dev_data)
test_dataset = Dataset.from_pandas(test_data)


dataset = DatasetDict({'train':train_dataset, 'dev':dev_dataset, 'test':test_dataset})
```
